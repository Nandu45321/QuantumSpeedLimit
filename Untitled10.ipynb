```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

omega_t = np.linspace(0, 2 * np.pi, 1000)

# Different theta values (angle of initial state from z-axis)
thetas = [np.pi/6, np.pi/4, np.pi/3, np.pi/2, 2*np.pi/3]
labels = [r'$\theta = \pi/6$', r'$\theta = \pi/4$', r'$\theta = \pi/3$',
          r'$\theta = \pi/2$', r'$\theta = 2\pi/3$']

fig, ax = plt.subplots(figsize=(8, 5))

for theta, label in zip(thetas, labels):
    # |<psi(t0)|psi(t)>|^2 = cos^2(wt/2) + cos^2(theta)*sin^2(wt/2)
    inner_product_sq = (np.cos(omega_t / 2)**2
                        + np.cos(theta)**2 * np.sin(omega_t / 2)**2)

    alpha = np.arccos(np.sqrt(inner_product_sq))

    ax.plot(omega_t, alpha, label=label)

# MT bound for theta = pi/2: alpha <= omega_t / 2
ax.plot(omega_t, omega_t / 2, 'k--', linewidth=2, label='MT bound ($\\theta=\\pi/2$)')
ax.fill_between(omega_t, omega_t / 2, np.pi, color='gray', alpha=0.2, label='Forbidden region')

# X axis ticks
x_ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
x_labels = ['$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels, fontsize=12)

# Y axis ticks
y_ticks = [0, np.pi/6, np.pi/3, np.pi/2, 2*np.pi/3, 5*np.pi/6, np.pi]
y_labels = ['$0$', r'$\frac{\pi}{6}$', r'$\frac{\pi}{3}$', r'$\frac{\pi}{2}$',
            r'$\frac{2\pi}{3}$', r'$\frac{5\pi}{6}$', r'$\pi$']


ax.set_yticks(y_ticks)
ax.set_yticklabels(y_labels, fontsize=12)
ax.set_xlabel(r'$\omega_0 t$', fontsize=13)
ax.set_ylabel(r'$\alpha(t)$', fontsize=13)
ax.set_title('Generalized MT Bound: $\\alpha$ vs $\\omega_0 t$', fontsize=13)
ax.legend()
plt.tight_layout()
plt.show()
```
